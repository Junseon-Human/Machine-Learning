```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

delta = np.random.rand(100, 1)
X = sorted(10 * np.random.rand(100, 1)) + delta
Y = sorted(10 * np.random.rand(100))

plt.plot(X, Y, "o")
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print(f"Linear regression coeff : {model.coef_}")  # 기울기
print(f"Linear regression intercept : {model.intercept_}")  # 절편

print(f"training Data evaluation : {model.score(X_train, y_train)}")
print(f"test Data evalution : {model.score(X_test, y_test)}")
# score : 주어진 데이터에 대한 모델의 예측 정확성이나 설명력을 측정
```

```python
forecast = model.predict(X_test)

plt.scatter(X_test, y_test)
plt.plot(X_test, forecast, "--y")
plt.show()
```

```python
forecast = model.predict(X_test)

plt.scatter(X, Y)
plt.plot(X_test, forecast, "--b")
plt.show()
```

```python
# 두 데이터를 하나로 봄
# 예측 수행
forecast = model.predict(X_test)


# 전체 데이터 분포 그리기 (회색 점)
plt.scatter(X, Y, label="Data")

# 테스트 데이터 분포 그리기 (파란 점)
plt.scatter(X_test, y_test, label="Test Data")

# 예측된 회귀선 그리기 (노란 점선)
plt.plot(X_test, forecast, "--y", label="Predict Line")

# 레이블 및 출력
plt.legend()
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
# Calculate MSE and RMSE
from sklearn.metrics import mean_squared_error
import numpy as np

y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
train_mse = mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)
train_rmse = np.sqrt(train_mse)
test_rmse = np.sqrt(test_mse)

print("Training Data MSE : {:.4f}".format(train_mse))
print("Test Data MSE : {:.4f}".format(test_mse))
print("Training Data RMSE : {:.4f}".format(train_rmse))
print("Test Data RMSE : {:.4f}".format(test_rmse))
```
