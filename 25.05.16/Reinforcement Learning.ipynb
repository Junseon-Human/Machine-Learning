```python
import gymnasium as gym
import sys



# 환경생성
env = gym.make("FrozenLake-v1", map_name="4x4", is_slippery=False)

# 초기화
state, info = env.reset()



# MACROS: 방향정의
LEFT = 0
DOWN = 1
RIGHT = 2
UP = 3



# 키보드입력매핑
arrow_keys = {"w": UP, "s": DOWN, "d": RIGHT, "a": LEFT}



# 상태 출력 함수 (4x4 grid)
def print_state(state):
    state_desc = env.unwrapped.desc.tolist()  # 현재 맵을 리스트로 만들어 반환함
    # 차원 리스트를 1차원으로 풀어 바이트 문자열을 일반 문자열로 바꿈
    state_desc = [[c.decode("utf-8") for c in line] for line in state_desc]
    row, col = divmod(state, 4)
    state_desc[row][col] = "P"
    for line in state_desc:
        print("".join(line))
    print()



print("Use w,a,s,d keys to move the agent")
print("Press 'q' to quit the game.\n ")

print("Initial state:")
print_state(state)

while True:
    key = input("Enter action (w,a,s,d) : ").strip().lower()
    if key == "q":
        print("Game aborted.")
        break

    if key not in arrow_keys:
        print("Invalid key. User w,a,s,d")
        continue

    action = arrow_keys[key]
    next_state, reward, terminated, truncated, info = env.step(action)

    action_names = {UP: "UP", DOWN: "DOWN", LEFT: "LEFT", RIGHT: "RIGHT"}
    print(
        f"\nAction: {action_names[action]}, State : {next_state}, Reward : {reward}, info : {info}"
    )

    print("Current state :")
    print_state(next_state)
    sys.stdout.flush()

    state = next_state

    if terminated or truncated:
        if reward == 1.0:
            print("Congratulations! You reached the goal!")
        else:
            print("Game over. You fell into a hole.")
        break
```
